## Problem 1: Local Search on the Ackley Surface
---

As discussed in class with Local Search problems the path to the goal is not as important as the goal itself. In this question we work with the Ackley function, a non-convex function typically used as a test for optimization algorithms.

The Achkey function has many local optima but only one global optimum that is $f(0,0) = 0$. Interestingly enough it might not be that easy to find this solution.

The Ackley function is defined as:

$$f(x,y):= -20exp[-0.2\sqrt{0.5(x^2 + y^2)}] -exp[0.5(cos(2\pi x) + cos(2\pi y))] + e + 20$$

````python
import numpy as np
from numpy import arange
from numpy import exp
from numpy import sqrt
from numpy import cos
from numpy import e
from numpy import pi
from numpy import meshgrid
from numpy import random
from numpy.random import randn
from numpy.random import rand
from numpy.random import seed
from numpy.random import uniform
# hill climbing search of the ackley objective function
from numpy import asarray
import plotly.graph_objects as go

# objective function
def f(x, y):
	return -20.0 * exp(-0.2 * sqrt(0.5 * (x**2 + y**2))) - exp(0.5 * (cos(2 * pi * x) + cos(2 * pi * y))) + e + 20

def create_ackley_figure(solution_points=[], soln_names=['A','B,','C']):
  """
  :param solution_points A numpy array of dimensions [i, n, 2]
  where i is the number of solution points you want to plot and n is the 
  number of steps used in estimating your solution. In each case use the same 
  number of steps if you want to plot all solutions on the same figure.

  :return An interactive Ackley function figure with/without solution points.
  """
  # define range for input
  r_min, r_max = -5.0, 5.0
  # sample input range uniformly at 0.1 increments
  xaxis = arange(r_min, r_max, 0.1) # (100,)
  yaxis = arange(r_min, r_max, 0.1) # (100,)
  # create a mesh from the axis
  x, y = meshgrid(xaxis, yaxis)

  # compute targets
  results = f(x, y)

  figure = go.Figure(data=[go.Surface(z=results, x=x, y=y,colorscale='Blues',showscale=False)])
  figure.update_layout(title='Ackley Function', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
  
  if len(solution_points):
    soln_colors = ['Rainbow', 'Reds', 'Viridis']
    for i in range(len(solution_points)):
      x_sln = np.array(solution_points[i])[:,0]
      y_sln = np.array(solution_points[i])[:,1]
      zdata = f(x_sln, y_sln)
      figure.add_scatter3d(name=soln_names[i], x=x_sln, y=y_sln, z=zdata, mode='markers',
      marker=dict(size=1, color=[i / len(x_sln) for i in range(len(x_sln))], colorscale=soln_colors[i]))
      
      figure.update_layout(showlegend=True)
      
  return figure
````

### General Function Definitions

````python
# objective function
def objective(v):
  """
  This function defines the Ackley surface, it has a minimum of 0. This 
  can be used to test if we have found points that yield the minimum value.
  :param v, a tuple representing a 2D point being tested.

  returns a value in the range [-5.0, 5.0]
  """
  x, y = v
  return -20.0 * exp(-0.2 * sqrt(0.5 * (x**2 + y**2))) - exp(0.5 * (cos(2 * pi * x) + cos(2 * pi * y))) + e + 20

  
# check if a point is within the bounds of the search
def in_bounds(point, bounds=asarray([[-5.0, 5.0], [-5.0, 5.0]])):
  """
  It is possible our optimzation method could move us outside of our search 
  space, so it is helpful to check.
  :param point a tuple representing a 2D point
  :param a 2D array, that describes the bounds of each dimension of our point.
  returns a Boolean of whether the point lies in the bounds or not.
  """
	# enumerate all dimensions of the point
  for d in range(len(bounds)):
    # check if out of bounds for this dimension
    if point[d] < bounds[d, 0] or point[d] > bounds[d, 1]:
      return False
  return True
````

````python
# Code for generating starting points
def generate_starting_points(bounds):
    start_position = rand(2)
    start_position[0] = start_position[0] * (bounds[0][-1] - bounds[0][0]) + bounds[0][0]
    start_position[1] = start_position[1] * (bounds[1][-1] - bounds[1][0]) + bounds[1][0]
    return start_position

# Code for generating stochastic neighbors
def generate_stochastic_neighbor(current_point, bounds, step_size):
    return current_point + random.uniform(low = -1, high = 1, size = len(bounds)) * step_size
````

````python
# Stochastic Hill-climbing with Random Restarts
#   stochastic_hillclimbing() is ran multiple times in random_restarts(), each time with a random starting point

def stochastic_hillclimbing(objective, bounds, n_iterations, step_size, start_pt):
  """
  :param objective, the function we are trying to optimize
  :param bounds, the boundaries of the problem
  :param n_iterations number of times to repeat stochastic hill climbing
  :param step_size how much we should move
  :param start_pt the point we start optimizing from.
  returns [solution, solution_value, candidates]
  """
  ### YOUR CODE HERE ###

  """
  This is the base stochastic hillclimbing algorithm intended to be ran multiple times in the below random_restarts() function

  Basic Idea: Start at start_pt, generate neighbors (based on step_size and bounds), then move to these neighbors to increase our "score" via the objective function

  Stochastic Hill-Climbing Algorithm:
    - Start from a determined position (start_pt). This position has a 'score' determined by the objective function (objective())
    - We want a list of positions moved to throughout this function for our Ackley Function to display
    - For n_iterations:
      - Generate a potential position to move to (generate_stochastic_neighbor())
      - Check if this "neighbor" is in bounds (determined by 'bounds' parameter)
      - If in bounds (use in_bounds()):
        - If neighbor's score is less than the current score (i.e. we want to minimize the objective function)
          - "Move" to this point by setting our position and score to these new values. Add this position to our list of candidates
  """
  # Function variables
  current_position = start_pt
  current_score   = objective(current_position)
  candidates = [current_position]                     # candidates = points we've reached in our loop

  # We want to iterate as many times as n_iterations allows
  for i in range(n_iterations):
    # Generate a stochastic neighbor -- potential position we could move to
    neighbor = generate_stochastic_neighbor(current_position, bounds, step_size)
    # if this neighbor is in bounds, run it through the objective function and see if this is better than our overall score. If so, this becomes our new position
    if in_bounds(neighbor, bounds):
       neighbor_score = objective(neighbor)
       if neighbor_score < current_score:
          current_position = neighbor
          current_score = neighbor_score
          # add this to our list of candidates
          candidates.append(current_position)
                
  return current_position, current_score, candidates

# random_restarts() takes the stochastic_hillclimbing() function and runs it multiple times, each time giving a different random start point
def random_restarts(objective, bounds, n_iter, step_size, n_restarts):
  """
  :param objective, the function we are trying to optimize
  :param bounds, the boundaries of the problem
  :param n_iter number of times to repeat stochastic hill climbing 
  :param step_size how much we should move
  :param n_restarts, the number of times we restart.
  returns [best solution, best solution value, visited points]
  """
  ### YOUR CODE HERE ###

  """
  This is fairly straightforward: we want to run the stochastic_hillclimbing() function n_iter times for n_restarts, where each restart gives a 
  randomly-generated starting position.

  Random Restarts:
    - We want to return our best solution, best solution value (best score), and all visited points. Set initial values for these
    - 
  """
  best_position  = None       # this gave issues when I initially set to 0
  best_score     = float('inf')   # We want to minimize this as much as possible
  visited_points = []

  for i in range(n_restarts):
    start_pt = generate_starting_points(bounds)
    position, score, candidates = stochastic_hillclimbing(objective, bounds, n_iter, step_size, start_pt)
    # add all new candidates to our list of visited points
    visited_points.extend(candidates)
    # we want to return the best position and score
    if score < best_score:
       best_position = position
       best_score    = score

  return best_position, best_score, np.array(visited_points)   # converted to np.array for Ackley function
  # return best_position, best_score, visited_points


# TEST CASE:

# seed the pseudorandom number generator
seed(240)
# define range for input
bounds = asarray([[-5.0, 5.0], [-5.0, 5.0]])
# define the total iterations
n_iter = 100
# define the maximum step size
step_size = 0.05
# total number of random restarts
n_restarts = 300
# perform the hill climbing search
best, score, shcr_candidates = random_restarts(objective, bounds, n_iter, step_size, n_restarts)
print('Done!')
print('f(%s) = %f' % (best, score))
````

### Part 2) Simulated Annealing

1.   Implement SA in the cell below.
2.   Visualize and comment on the candidate points visited.

````python
def get_temperature_schedule(epoch, temp, cos_frequency, diff):
  """
  :param epoch Iterations elapsed
  :param temp Temperature
  :param cos_frequency Cosine annealing cycle frequency
  :param diff Difference between value of candidate solution and current solution.
  """
  # calculate temperature for current epoch (1.001 to prevent t from going to zero)
  t = temp * (1.001 + np.cos(epoch*(2*np.pi*cos_frequency)))/2
  # calculate acceptance criterion
  criterion = exp(-diff / t)
  return criterion

# simulated annealing algorithm
def simulated_annealing(objective, bounds, n_iterations, step_size, temp, cos_frequency):
  """
  :param objective, the function we are trying to optimize
  :param bounds, the boundaries of the problem
  :param n_iterations number of times to repeat stochastic hill climbing
  :param step_size how much we should move
  :param temp temperature for each epoch
  :param cos_frequency Cosine annealing cycle frequency
  returns [solution, solution_value, candidates]
  """
  ### YOUR CODE HERE ###

  """
  Simulated Annealing - occasionally going uphill to try and break away from a local minimum

  Idea:
    - We want to keep track of our current and best position and score throughout this function
    - We want to keep track of our list of candidates as before
    - For n_iterations:
      - Generate a stochastic neighbor and, if the neighbor is in bounds, take its score
      - The temperature schedule calculation requires the difference between neighbor score and current score, so we
          take this calculation down now
      - Calculate the temperature t
      - Calculate the probability of accepting a worse successor (prob)
      - If our neighbor score is less than our current score, or if we pass the threshold to accept a worse successor:
        - Update our current values to reflect the neighbor values (we move to new position)
        - Add this new position to our list of candidates we want to return
        - Check if these new values are better than our best-tracked values. If so, set new best values
  """

  current_position = generate_starting_points(bounds)
  current_score    = objective(current_position)
  best_position = current_position
  best_score    = current_score
  candidates = [current_position]

  for i in range(n_iterations):
    # As with hillclimbing we generate a stochastic neighbor -- potential position we could move to
    neighbor = generate_stochastic_neighbor(current_position, bounds, step_size)
    # if this neighbor is in bounds, run it through the objective function and see if this is better than our overall score. If so, this becomes our new position
    if in_bounds(neighbor, bounds):
       neighbor_score = objective(neighbor)
       diff = neighbor_score - current_score  # needed for temperature calculation
    
    # calculate the temperature and our probability that we move 'despite' new position moving uphill
    temperature = get_temperature_schedule(i, temp, cos_frequency, diff)
    prob = np.random.rand() < np.exp(-diff / temperature)

    """
    We accept the new position as the best solution if:
      1) the new position is better (new score < current best score)
      2) If the new position is worse and our temperature-dependent probability (prob) returns a true value
    """

    if diff < 0 or prob:
       current_position = neighbor
       current_score    = neighbor_score
       candidates.append(current_position)

       # Finally we update our best position if this new position is indeed better
       if current_score < best_score:
          best_position = current_position
          best_score    = current_score
    
  return best_position, best_score, np.array(candidates)    # setting to np.array since this worked with Part 1

# seed the pseudorandom number generator
seed(240)
# define range for input
bounds = asarray([[-5.0, 5.0], [-5.0, 5.0]])
# define the total iterations
n_iterations = 10000
# define the maximum step size
step_size = 0.1
# initial temperature
temp = 5
# cosine annealing cycle frequency
cos_frequency = .01
# perform the simulated annealing search
best, score, sa_candidates = simulated_annealing(objective, bounds, n_iterations, step_size, temp, cos_frequency)
print('Done!')
print('f(%s) = %f' % (best, score))
````
